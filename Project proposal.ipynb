## Manager's section
The focus of this project will be to see if we can make a model able to make recommendations based on a set of songs. 
 
To do so, we will have to prepare the data and identify what factors are the most impactful. This will be challenging since the data has many factors (speechless, liveness, temp etc.), so isolating/testing the impact of different factors, either isolated or combined will take time. Our model is also ambitious: instead of simply coming up with a good model that can predict the likelihood of a song being popular, we chose to try to set a model that can recommend songs. This will be challenging, as we have limited knowledge of recommendation models. Moreover, measuring success will be difficult, as the accuracy of a recommendation is hard to quantify.  

Team members were assigned their duties and responsibilities based on the part of the project they were the most interested in. Anne and Andrew will focus on supporting Samuel by trying to identify which factors are the most impactful and should be used by Samuel's model. Samuel will focus on finding and training our recommendation model. All three of us are responsible for generating visual content (graphs and slides) explaining our process and our results.  

To consider ourselves successful, we want to be able to: explain how we came up with our model, as well as see if it can recommend popular songs accurately (if we feed it a set of popular songs, we should get back songs that have similar features).
